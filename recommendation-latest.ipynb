Since the mean of meadian of ingredients are around 18, we will randomly gengerate 17 ingredients and recommend 1 ingredient.<br>

We will recommend new ingredients by finding the most similar recipe in existing recipes and recommend the ingredients that this most similar recipe has. Then we will calculate a recommendation index for the recommeneded ingredients. For each recommended ingredient, we count how many times it occurs together in a recipe with each of the 17 ingredients and add those 17 numbers up. We will use this value as the index of to which extent we recommend a ingredient. The larger the index is, the more we will recommend a ingredient as this indicates that the recommended ingredient occurs simultaneously with the 17 ingredients we have more frequently and thus are more correlated with those 17 ingredients.
